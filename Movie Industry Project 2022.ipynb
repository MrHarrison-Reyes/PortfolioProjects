# Final correlation check
## hypothesis: Company and budget will have high correlation

```python
# scatter plot with budget vs gross to test correlation hypothesis
plt.scatter(x=df['budget'], y=df['gross']) #x axis is budget and y axis is gross
plt.title('Budget v. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget For Films')
plt.show
```

```python
df.corr() #pearson method (default)
```

Shown in the correlation matrix above, budget and gross have the highest correlation at 0.740395. So far this confirms the original hypothesis.

```python
#visualize correlation matrix
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix (Numeric Features Only)')
plt.xlabel('Film Features')
plt.ylabel('File Features')
plt.show()
```

Correlation Matrix only currently features numeric values. Other string variables (i.e. company, director, country, etc.) will need to be converted in order to include them in the correlation analysis.

```python
#change column name from object to category for correlation matrix inclusion
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
#visualize correlation matrix including all columns
correlation_matrix=df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix (all fields)')
plt.xlabel('Film Features')
plt.ylabel('File Features')
plt.show()
```

```python
# Analysis: Budget and gross remain the highest correlation categories making the initial hypothesis correct.
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
